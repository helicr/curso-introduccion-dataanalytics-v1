# Listas
Las listas son una de las estructuras de datos más versátiles de Python. Contienen una colección de valores. Las listas pueden contener elementos del mismo tipo o de tipos diferentes, y los elementos también se pueden cambiar. Es fácil crear una lista simplemente definiendo una colección de valores separados por comas entre corchetes. Las líneas de código a continuación crean una lista, 'movie', que contiene los nombres de las películas más exitosas y sus colecciones de taquilla en miles de millones de dólares

```python
movie = ["Avengers: Endgame", 2.796, "Avatar", 2.789, "Titanic", 2.187, "Star Wars: The Force Awakens", 2.068, "Avengers: Infinity War", 2.048] 
type(movie)
```

```python
movie
```

```python
movie2 = [["Avengers: Endgame", 2.796], ["Avatar", 2.789], ["Titanic", 2.187], ["Star Wars: The Force Awakens", 2.068], ["Avengers: Infinity War", 2.048]] 
movie2
```

Crear subconjuntos de listas es fácil en Python, y comienza con el índice 0. Por eso se le llama indexación basada en cero. Supongamos que queremos ver el primer elemento de la lista 'movie'. Podemos hacerlo fácilmente con la sintaxis

```python
movie[0]
```

Al igual que los subconjuntos, la división de listas es fácil en Python. La sintaxis está en el formato "lists start: end", en el que el elemento 'start' es inclusivo pero el elemento 'end' es exclusivo. El siguiente ejemplo imprime la salida del tercer al quinto elemento, excluyendo el sexto.

```python
movie[3:6]
```

También podemos realizar la manipulación de la lista para cambiar, agregar o eliminar elementos de la lista, como se muestra a continuación.

```python
movie + ["Jurassic World", 1.67]
```

```python
tuple_1 = 20,30,40,50,60
tuple_1
```

# Diccionarios
Las listas son convenientes, pero no interactivas. El diccionario es una alternativa interactiva a las listas. La principal diferencia entre los dos es que las listas están indexadas por un rango de números, mientras que un diccionario está indexado por claves únicas que pueden usarse para crear tablas de búsqueda. Un par de llaves, {}, crea un diccionario vacío, como se muestra a continuación:

```python
mov = {"Avengers: Endgame": 2.796, "Avatar":2.789, "Titanic":2.187} 

mov["Avatar"]
```

# Arrays Numpy
Otra alternativa a las listas de Python son las matrices numpy, que son colecciones de puntos de datos. Las listas son poderosas, pero para la ciencia de datos, necesitamos una alternativa que tenga velocidad y permita operaciones matemáticas sobre los elementos. Los arrays permiten cálculos más simples, como se ilustra en el siguiente ejemplo.

```python
import numpy as np

runs = [100, 89, 75, 28]
np_runs = np.array(runs)
                     
over = [10, 9, 6, 2]
np_over = np.array(over)

runs_over = np_runs / np_over
runs_over
```

Es posible hacer subconjuntos de las matrices numpy de manera similar a las listas.

```python
runs_over[0]
```

También es posible crear matrices n-dimensions usando numpy. En el siguiente ejemplo, creamos matrices bidimensionales, que contienen corridas y más

```python
example_2d = np.array([[100, 89, 75, 28, 35], 
                   [10, 9, 6, 2, 4]]) 

example_2d
```

El subconjunto de las matrices numéricas n-dimensionas también sigue el método de indexación de base cero. Se muestran algunos ejemplos en las líneas de código a continuación

```python
# Primer elemento
print(example_2d[0])

# Primer y segundo elemento
print(example_2d[0][1])

# Segundo y tercer elemento y siguientes
example_2d[:,1:3] 
```

# Data Frames
La matrices que son colecciones de puntos de datos, son potentes para hacer calculos. Sin embargo, la limitación de una matriz es que solo puede manejar un tipo de datos. Pero para problemas de ciencia de datos del mundo real, necesitamos conjuntos de datos para manejar diferentes tipos de datos, como texto, float, entero, etc. La solución es 'Dataframes', que es el formato de datos de facto para el aprendizaje automático y el modelado predictivo. 
La estructura típica de un 'Dataframe' contiene observaciones en filas y variables en columnas. Se puede construir un dataframe usando el diccionario, como en las líneas de códigos a continuación. 
La primera línea de código a continuación importa la libreria pandas. La segunda línea crea el diccionario que contiene los valores almacenados en las variables 'movie', 'collections' y 'release_yr', respectivamente. La tercera línea convierte este diccionario en un dataframe pandas, mientras que la cuarta línea imprime el dataframe resultante.

```python
import pandas as pd 

dic_movie = { 
    "movie":["Avengers: Endgame", "Avatar", "Titanic", "Star Wars: The Force Awakens", "Avengers: Infinity War"], 
    "collections":[2.796, 2.789, 2.187, 2.068, 2.048],
    "release_yr":[2019, 2009, 1997, 2015, 2018]}

movie_df = pd.DataFrame(dic_movie)
movie_df
```
