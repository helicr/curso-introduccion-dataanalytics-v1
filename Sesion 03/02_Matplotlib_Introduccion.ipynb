## Componentes

### Los datos

Matplotlib representa series de datos almacenadas en variables de tipo lista de python:
- listas
- numpy arrays
- pandas series

```python
data_1 = np.random.normal(loc=1.0,scale=0.5, size=1000)
print(data_1.shape)
print(data_1.dtype)
print(data_1[:10])
```

### Elementos

Se accede a los elementos básicos con `plt`.

```python
plt.title("Esto es un titulo")

plt.xlim([0, 1])
plt.ylim([-2, 4])

plt.show()
```

## Graficos Univariable

### Histograma: Una variable continua

Los histogramas representan en altura la acumulación de casos en intervalos de valor. Una sola variable continua puede representarse de esta manera.

```python
plt.hist(data_1, bins=50)
plt.show()
```

```python
plt.rc("figure", figsize=(2.5,2.5))
```

```python
plt.hist(data_1, bins=50)
plt.show()
```

```python
plt.hist?
```

Para el ejemplo, crearemos varias distribuciones de datos, y observaremos como se muestran en histogramas.

```python
data_2 = np.random.normal(loc=1.5,scale=0.1, size=500)
data_3 = np.random.normal(loc=0.5,scale=0.6, size=500)
```

```python
plt.hist(data_2, bins=20, alpha=0.5, label="data 1", ec="black")
plt.hist(data_3, bins=20, alpha=0.5, label="data 3", ec="black")

plt.legend(loc='best')
plt.show()
```

```python
plt.legend?
```

### Barras: Una variable categórica

Para representar una variable categórica usamos el clásico diagrama de barras. Se representa en alturas (o longitudes) ciertas cantidades asociadas a ciertos valores.



```python
ciudades = ["Madrid", "Barcelona", "Sevilla", "Valencia"]
#freq = list(np.random.random_integers(1, 10, 4))
freq = list(np.random.randint(1, 10, 4))
x = list(range(len(ciudades)))
```

```python
print(ciudades)
print(x)
print(freq)
```

```python
plt.rc("figure", figsize=(5,5))
```

```python
plt.bar(x, freq, align="center", tick_label=ciudades, ec="black")
plt.show()
```

## Graficos Multivariable

### Scatter (variables numéricas sin relacion)

```python
data_4 = np.random.normal(loc=0.5,scale=2, size=500)
plt.scatter(data_2, data_4, alpha=0.5)
plt.title("Scatter Plot")
plt.show()
```

Para reprepresentar varios conjuntos de variables, por defecto, cambia los colores seguun un esquema propio.

Pero se pueden especificar diferencias especificas como:

- `marker` para la forma del item
- `color` para el color
- `s` para el tamaño

```python
plt.scatter(data_2, data_4, alpha=0.5, marker="*", color="green", s=5)
plt.scatter(data_2, data_3, alpha=0.5, marker="8", color="blue", s=10)
plt.show()
```

Tambien podemos combinar varias tipos en una grafica. Pero hay que tener cuidado con los ejes.

```python
sizes = np.round(np.abs(data_4)*20, 0)
```

```python
sizes[:20]
```

```python
plt.scatter(data_2, data_4, alpha=0.5, marker="8", color="green", s=sizes)
plt.scatter(data_2, data_3, alpha=0.5, marker="*", color="blue", s=5)
plt.show()
```

```python
bins = np.linspace(np.min(data_2), np.max(data_2), 250)
```

```python
plt.scatter(data_2, data_4, alpha=0.5, marker="8", color="green", s=sizes)
plt.scatter(data_2, data_3, alpha=0.5, marker="*", color="blue", s=5)
plt.hist(data_2, bins=bins, alpha=0.5, label = "hist data 2", ec="black")
plt.show()
```

### Grafico de Lineas (Plot)

También se puede unir la secuencia de puntos, con lineas mediante un plot

```python
x = np.linspace(0, 10, 10)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.show()
```

A la que se le pueden añadir otros elementos: Varias lineas, leyenda, valores...

https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html

```python
plt.plot(x, x, "o-", label="Lineal")
plt.plot(x, y, "x-", label="Cuadratica")

plt.legend(loc="best")

plt.title("Relacion lineal y cuadratica")

plt.xlabel("Eje x")
plt.ylabel("Eje y")

plt.show()
```
